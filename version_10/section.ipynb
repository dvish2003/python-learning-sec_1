## Generators

```python
def get_odd_numbers(upper_limit):
    odd =[]
    for i in range(0,upper_limit):
        if i % 2 == 1:
         odd.append(i)

    return odd


list = get_odd_numbers(20)
print(list)
```

```python
# odd numbers generate කරන්න generator function එකක් define කරමු
def get_odd_numbers(upper_limit):
    # 0 සිට upper_limit (exclusive) දක්වා loop එකක්
    for i in range(0, upper_limit):
        # number එක oddද බලමු (2ට බෙදා කොට remainder 1ද කියලා)
        if i % 2 == 1:
            # console එකට odd number එක print කරනවා
            print('odd ', i)
            # generator එක return කරනවා, එය iteration එකේදී yield කරනවා
            yield i

# generator function එක call කරනවා, list කියලා variable එකට assign කරනවා
# list කියලා assign කරන්නෙ iterator එකක් assign කරනවා, list එකක් නෙමෙයි
list = get_odd_numbers(10)

# generator එක iterate කරන loop එක
for i in list:
    # generator එකෙන් ලබා ගත් value එක print කරනවා
    print('loop', i)
```
