## Comprehension

```python
a = [12,23,32,121,12]
b = a  # not a copy a ma thami b ta assign wenne

b.append(55)

d = list(a)  # copy ekak wenne mehema
d.append(100)

print(b)
print(a)
print(d)
```

```python
##. copy with out list function
a = [12,23,32,121,12]

b =[]

for i in a:
    b.append(i)

b.append(99)

print(a)
print(b)

#list comprehension
c = [12,23,32,121,12]
d = [i for i in c] # c array eken eken ek item aragena d ta append karanawa
d.append(88)

print(c)
print(d)



```

```python
def check_odd_even(number):
    return number % 2 == 0


a = [12,23,32,121,12,55,66,77,88]
b =[check_odd_even(i) for i in a]
print(b)
```

```python
def check_odd_even(number):
    return "Odd" if number % 2 != 0 else "Even"


a = [12,23,32,121,12,55,66,77,88]
b =[check_odd_even(i) for i in a]
print(b)
```
