```python
##. copy with out list function
a = [12,23,32,121,12]

b =[]

for i in a:
    b.append(i)

b.append(99)

print(a)
print(b)

#list comprehension
c = [12,23,32,121,12]
d = [i for i in c] # c array eken eken ek item aragena d ta append karanawa
d.append(88)

print(c)
print(d)



```

```python
def check_odd_even(number):
    return "Odd" if number % 2 != 0 else "Even"


a = [12,23,32,121,12,55,66,77,88]
b =[check_odd_even(i) for i in a]
print(b)
```

## conditional comprehension

```python
def is_odd(number):
    return 'Odd' if number % 2 == 1 else 'Even'

x = [12,32,35,43,21,90,32,45,32,12,345,65]

y = [{value : is_odd(value)} for i,value in enumerate(x) if i % 2 == 0]
#index = i  value = value meken i index ek aragenalist eke ekak ara ekak gnnw. e kiynne i % 2 = 0 index ek ta adala value ek witrk aran pass krnw

print(x)
print(y)
```
