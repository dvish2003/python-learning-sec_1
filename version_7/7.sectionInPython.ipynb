## Break

```python
count_2 = 0

while True:
    if count_2 >= 25:
        break
    print(f"Count_2 is {count_2}")
    count_2 += 1
    
```

## Iteration Dictionary and Set

```python
d = {
    "vishan":178,
    "ajay":165,
    "ram":172,
    "siri":160
}

for name in d:
    height = d[name]
    print(name, height)


for name , height in d.items():
    print(name)
```

## Enumerate tuple

## Enumerate.  Dictionary

```python
d = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "occupation": "Engineer"
}
for i,x in enumerate(d.items()):
    name,height = x
    print(i,x)
    print(name,height)
```
