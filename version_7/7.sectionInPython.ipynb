## While Loop

```python
x = [12,32,44,43,13,67,99,9,23,21]
```

```python
for item in x:
    if item % 2 == 0:
        print(f"{item} is even")
    else:
        print(f"{item} is odd")
```

```python
y = []
```

```python
#decending order. 

#01
for item in x:
    y.append(item)
y.sort(reverse=True)
print(y)
    
```

```python
#While loop
count = 0
while count < 10:
    print(f"Count is {count}")
    count += 1
```

## Break And Continue

## Break

```python
count_2 = 0

while True:
    if count_2 >= 25:
        break
    print(f"Count_2 is {count_2}")
    count_2 += 1
    
```

```python
z =[5,3,8,6,7,2,9,1,4,0,32,34,55,21,34,56,65,235,678,76,89,90]

for item in z:
    if item >= 100:
        break
    print(item)
```
