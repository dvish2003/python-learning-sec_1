## Funtions

```python
mark = int(input("Enter the numerical grade (0-100): "))
def get_grade(grade):
    if 90 <= grade <= 100:
        return 'A'
    elif 80 <= grade < 90:
        return 'B'
    elif 70 <= grade < 80:
        return 'C'
    elif 60 <= grade < 70:  
        return 'D'
    elif 0 <= grade < 60:
        return 'F'
    else:
        return 'Invalid grade'
# This function takes a numerical grade (0-100) and returns the corresponding letter grade.

get_grade(mark)
    

```
