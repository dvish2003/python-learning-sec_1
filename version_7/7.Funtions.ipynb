## Funtions

```python
mark = int(input("Enter the numerical grade (0-100): "))
def get_grade(grade):
    if 90 <= grade <= 100:
        return 'A'
    elif 80 <= grade < 90:
        return 'B'
    elif 70 <= grade < 80:
        return 'C'
    elif 60 <= grade < 70:  
        return 'D'
    elif 0 <= grade < 60:
        return 'F'
    else:
        return 'Invalid grade'
# This function takes a numerical grade (0-100) and returns the corresponding letter grade.

get_grade(mark)
    

```

## Default Values

```python
def get_grade_1(marks , subject="Unknown"):     
    print(f"The grade for {subject} is: {marks}")


get_grade_1(21, "Mathematics")
get_grade_1(85)


# python rule ekak tami default value thina para meter ekkt passe thina okkom para meter wlt default values thinn oni

def get_grade_2(marks, subject="Unknown", year=2023):
    print(f"The grade for {subject} in the year {year} is: {marks}")

get_grade_2(95, "Physics", 2022)
get_grade_2(88, "Chemistry")  # year will default to 202
get_grade_2(76)  # subject will default to "Unknown" and year to 2023
```
