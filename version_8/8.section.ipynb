## Packed Argument

```python
def get_grade(*marks): #packed argument used here
     print(type(marks))

     total = 0
     for mark in marks:
         total += mark  
     print(f"Total marks: {total}")

get_grade(45, 78, 89, 90)
```

## keyword argument

```python
def my_form(**params):
    print(type(params))
    print(params)
    
my_form(name="Vishan", age=25, country="Sri Lanka")
```

```python
def my_form(**params):   #its like a dictionary
    if "name" not in params:
        print('error: name not provided')
    else:
        print("hello", params["name"])
my_form(name="Vishan", age=25, country="Sri Lanka")
my_form(age=25, country="Sri Lanka")
```

## Packed arguments + Keyword Argument

```python
def my_form(*values, **params):
    print("values:", values)
    print("params:", params)

my_form(1,2,3, name="Vishan", age=25, country="Sri Lanka")
```

# dict convert name argument

```python
def my_form(name , height):
   print(name , height)

args = {"name": "Vishan", "height": 175}
my_form(**args)
```

## return

```python
def get_grade(grade):
    if 90 <= grade <= 100:
        return 'A'
    elif 80 <= grade < 90:
        return 'B'
    elif 70 <= grade < 80:
        return 'C'
    elif 60 <= grade < 70:  
        return 'D'
    elif 0 <= grade < 60:
        return 'F'
    else:
        return 'Invalid grade'

result = get_grade(85)
print(f"The grade is: {result}")
```

```python
def get_grade(mark,subject="unknown"):

    if mark < 0 or mark > 100:
        grade = None
    elif mark >= 90:
        grade = 'A'
    elif mark >= 80:    
        grade = 'B'
    elif mark >= 70:
        grade = 'C'
    elif mark >= 60:
        grade = 'D'
    else:
        grade = 'F'
    
    return subject, grade


subject,grade = get_grade(92, subject="Math")
print(type(grade))
print(grade)
print(subject)


```
